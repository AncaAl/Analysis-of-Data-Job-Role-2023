```python
import pandas as pd
import pycountry                    # library to convert country codes
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np


# Load dataset from an online source
url = 'https://raw.githubusercontent.com/AncaAl/Project-3/main/ds_salaries.csv'
df = pd.read_csv(url)
```

```python
#change the name of the columns
df.rename(columns = {"company_location" : "country_code2", "work_year" : "year"}, inplace=True)
```

```python
# Function to convert a two-letter code to a full country name using pycountry
def convert_code_to_name(code):
    try:
        country = pycountry.countries.get(alpha_2=code)     # Lookup the country by alpha_2 code
        return country.name
    except KeyError:
        return code
df['country'] = df['country_code2'].apply(convert_code_to_name)
def convert_alpha2_to_alpha3(code):
    try:
        alpha_3 = pycountry.countries.get(alpha_2=code).alpha_3
        return alpha_3
    except KeyError:
        return code
df['country_code'] = df['country_code2'].apply(convert_alpha2_to_alpha3)
df
```

```python
df.rename(columns = {"country_code" : "CountryCode", "work_year" : "year"}, inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator
last_10_years_df = pd.DataFrame(merged_dataframe)

# Add inflation data for 2023 for USA and CAN
new_data = [
    {'year': 2023, 'inflation_rate': 4.1, 'CountryCode': 'USA'},
    {'year': 2023, 'inflation_rate': 3.4, 'CountryCode': 'CAN'},
    {"year": 2023, 'inflation_rate': 5.7, 'CountryCode': 'FRA'},
    {'year': 2023, 'inflation_rate': 4.4, 'CountryCode': 'IND'},
    {"year": 2023, 'inflation_rate': 4.1, 'CountryCode': 'CHN'},
    {'year': 2023, 'inflation_rate': 5.2, 'CountryCode': 'GBR'},
    {'year': 2023, 'inflation_rate': 3.4, 'CountryCode': 'ESP'},
    {'year': 2023, 'inflation_rate': 5.9, 'CountryCode': 'DEU'},

]
new_data_df = pd.DataFrame(new_data)

last_10_years_df = pd.concat([last_10_years_df, new_data_df], ignore_index=True)

# Check the updated DataFrame
print(last_10_years_df)

# Plotting the bar chart for inflation rate over time by country
plt.figure(figsize=(14, 8))

# Plotting bar chart for inflation rate over time by country
for country in last_10_years_df['CountryCode'].unique():
    country_data = last_10_years_df[last_10_years_df['CountryCode'] == country]
    plt.bar(country_data['year'], country_data['inflation_rate'], label=country, alpha=0.7)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
# Adding labels and title
plt.title('Inflation Rate Over the Last 4 Years by Country')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.legend(title='CountryCode', bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(axis='y')  # Adding grid on y-axis only for bar chart
plt.tight_layout()
plt.show()
```
